```python
# identify high-cardinality columns (columns with many unique values).
#df.nunique().sort_values(ascending=True)
df.nunique().sort_index(ascending=False)
```

```python
# Remove unnecessary columns.
# PassengerId, Can the number 1, 2, or 3 help predict survival? It's just an identifier. The model cannot learn any meaningful pattern from it.
df = df.drop(columns='PassengerId')
```

```python
df.columns
```

```python
# Extracting Title from Name.
df['Name'].head(20)
```

```python
#"I'm not deleting this column."
# "I'm stealing the useful part."
# This is called:
# Feature Engineering"I'm not deleting this column."
# "I'm stealing the useful part."
# This is called:
# Feature Engineerin
# Our First Engineered Feature
df['Title'] = df['Name'].str.extract(r'([A-Za-z]+)\.', expand=False)
```

```python
df[['Name','title']].head(20)
# The raw Name column isn't directly usable by the model, but it contains meaningful information such as passenger titles. I extracted the title because it may capture demographic information like age, gender, or social status, which could improve prediction.
```

```python
df['Title'].value_counts()
```

```python
# Instead of treating them separately,
# we combine them:
rare_titles = [
    'Lady', 'Countess', 'Capt', 'Col',
    'Don', 'Dr', 'Major', 'Rev',
    'Sir', 'Jonkheer', 'Dona'
]
df['Title'] = df['Title'].replace(rare_titles, 'Rare')
```

```python
df['Title'].value_counts()
```

```python
# create one new feature for family
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df[['SibSp', 'Parch', 'FamilySize']].head(20)
```

```python
# Handle Missing Values
df.isnull().sum()
```

```python
df['Age'].head(20)
```

```python
# Age usually has some skew and outliers, so median is a safe choice.
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Embarked
# Only 2 missing value
# Fill with the most frequent value (mode).
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode())
```

```python
df['Embarked'].head()
```

```python
# Column 3: Cabin
# 687 missing values, Should we fill 687 values?
# More than 75% of the column is missing.
# Fill Probably not. Imagine replacing 687 missing cabins with a guessed value it would introduce a lot of noise.
# option 1 Drop the column. option-2  Extract the deck letter (A, B, C, etc.) before deciding.
# will choose option A
df = df.drop(columns='Cabin')
```
